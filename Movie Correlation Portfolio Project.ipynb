```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import copy

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data
df = pd.read_csv(r'D:\Users\charl\Downloads\movies.csv')

```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

#Chadf.dtypes
```

```python
# Change null values to 0 relevant columns

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)

df['votes'] = df['votes'].fillna(0)

df['runtime'] = df['runtime'].fillna(0)
```

```python
# Change data types of relevant columns from float to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Organize data by most lucrative to least

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

df.drop_duplicates();
```

```python
# How does budget correlate with revenue? Other relationships?
```

```python
# Build scatterplot -- budget vs. gross
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget For Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross using Seaborn, making adjustments for readability

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Examine correlations for all columns that include numeric data types

df.corr(method='pearson') # Pearson, Kendall, Spearman each have different methodology

# High correlation between budget and gross established!
```

```python
# Visualize correlations alternative

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
